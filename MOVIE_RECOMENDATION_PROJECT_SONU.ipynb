```python
import numpy as np
import pandas as pd
```

```python
cred=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/credits.csv')
cred.head()
```

```python
movies= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movies.csv')
movies.head()
```

```python
df= cred.merge(movies,on='title')
```

```python
# selecting relevant feature
features= ['genres','keywords','tagline','cast','crew']
```

```python
#replacing null values
for feature in features:
  df[feature]=df[feature].fillna(' ')
```

```python
df.isnull().sum()
```

```python
#our combined features are taking into new variable data

data= df['genres']+" "+df['keywords']+' '+df['tagline']+''+df['cast']+''+df['crew']
```

```python
data
```

```python
#converting tect to vector
from sklearn.feature_extraction.text import TfidfVectorizer
vect= TfidfVectorizer()
```

```python
vectorised= vect.fit_transform(data)
```

cosine similiarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectorised)
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie = input('ENTER THE MOVIE NAME TO GET SIMILAR MOVIES :')
```

```python
# getting all movie names into a list from the data set
all_movies = df['title'].tolist()
```

```python
# close match with the movie given by user
import difflib
get_close_match = difflib.get_close_matches(movie,all_movies)
print(get_close_match)
```

```python
close_match= get_close_match[0]
close_match
```

```python
# finding index of movie based on titile
index_of_movie = df[df.title == close_match].index.tolist()
index_of_movie[0]
```

```python
#getting similar movies
similarity_score = list(enumerate(similarity[index_of_movie[0]]))
similarity_score
```

```python
#sorting movies based on similarity score

similar_movies = sorted(similarity_score,key= lambda x: x[1],reverse=True)
print(similar_movies)
```

```python
#similar movies
print('SIMILAR MOVIES RELATED TO ',movie ,'ARE LISTED BELOW \n' )
i=1
for mov in similar_movies:
  index=mov[0]
  title_index= df[df.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_index)
    i+=1
```
